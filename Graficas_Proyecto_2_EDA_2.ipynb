```python
import matplotlib.pyplot as plt
import csv
```

```python
entradas_paralelo = []
entradas_serial = []
tiempos_paralelo = []
tiempos_serial = []

with open('datos_paralelo.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        entradas_paralelo.append(int(row[0]))
        tiempos_paralelo.append(float(row[1]))

with open('datos_serial.csv') as f:
    reader = csv.reader(f)
    for row in reader:
      entradas_serial.append(int(row[0]))
      tiempos_serial.append(float(row[1]))
```

Para ver como supera el paralelo al serial

```python
e_paralelo = []
e_serial = []
t_paralelo = []
t_serial = []
with open("datos_serial_2.csv") as f:
  reader = csv.reader(f)
  for row in reader:
    if int(row[0])%50 == 0:
      e_serial.append(int(row[0]))
      t_serial.append(float(row[1]))

with open("datos_paralelo_2.csv")as f:
  reader = csv.reader(f)
  for row in reader:
    if int(row[0])%50 == 0:
      e_paralelo.append(int(row[0]))
      t_paralelo.append(float(row[1]))
```

```python
fig,ax = plt.subplots()
ax.plot(entradas_paralelo,tiempos_paralelo,label = 'paralelo',color='m')
ax.plot(entradas_serial,tiempos_serial,label='serial',color='#08E580')
ax.set_xlabel('Tamano de entrada')
ax.set_ylabel('Tiempos de ejecucion')
ax.legend(loc='best')
plt.savefig('todos_los_datos.png')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(e_paralelo,t_paralelo,label = 'Paralelo',color='m')
ax.plot(e_serial,t_serial,label='Serial',color='#08E580')
ax.set_xlabel('Tamano de entrada')
ax.set_ylabel('Tiempos de ejecucion')
ax.legend(loc='best')
plt.savefig('primeros_200.png')
plt.show()
```
